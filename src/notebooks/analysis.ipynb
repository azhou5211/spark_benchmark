```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(rc={'figure.figsize':(15,10)})
```

```python
df = pd.read_excel("../data/Running_time.xls", sheet_name="Sheet2")
df
```

```python
df['Data'] = df['Data'].astype('category')
df['Workers'] = df['Workers'].astype('category')
```

```python
fig, ax = plt.subplots()
sns.lineplot(data=df, x="Data", y="Time(seconds)", hue="Workers", style="Cloud", markers=True, ax=ax)
ax.set_xlabel("Number of Data Points (Note: In Millions)")
ax.set_ylabel("Time to Compute K-Means (seconds)")
ax.set_title("Running time of K-Means vs Number of Data Points")
plt.xticks([0, 1044822, 5139675, 10434586, 15575834], ['0', '1M', '5M', '10M', '15M'])
plt.show()
```
